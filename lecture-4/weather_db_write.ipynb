### DBの作成テスト

```python
import sqlite3
import os

# 今いる場所にDBを作る
db_path = os.path.join(os.getcwd(), 'weather_forecast.db')
print(f"作成予定のパス: {db_path}")

try:
    conn = sqlite3.connect(db_path)
    cur = conn.cursor()
    cur.execute("CREATE TABLE IF NOT EXISTS test (id INTEGER)")
    conn.commit()
    conn.close()
    print("成功：ファイルが作成されました！")
except Exception as e:
    print(f"エラー発生: {e}")
```
