```python
x=1 
y=0
```

```python
x/y
```

```python
try:
    result = x / y
except ZeroDivisionError as e:
    result = None
    print("エラーが発生しました:", e)
    print("0で割ることはできません。")

print("結果:", result)
```

###
エラーが発生しなかった場合のみ実行したい場合

```python
try:
    result = x / y
except ZeroDivisionError as e:
    result = None
    print("エラーが発生しました:", e)
    print("0で割ることはできません。")
except Exception as e:
    result = None
    print("エラーが発生しました:", e)
    print("変数が定義されていません。", e)
    
print("結果:", result)
```

```python
z=3

try:
    result = x / z
except ZeroDivisionError as e:
    result = None
    print("エラーが発生しました:", e)
    print("0で割ることはできません。")
except Exception as e:
    result = None
    print("エラーが発生しました:", e)
    print("変数が定義されていません。", e)
else:
    print("エラーは発生しませんでした。")
    
print("結果:", result)
```

## エラーの発生の有無に関わらず何かを実行したい

```python
z=3

try:
    result = x / z
except ZeroDivisionError as e:
    result = None
    print("エラーが発生しました:", e)
    print("0で割ることはできません。")
except Exception as e:
    result = None
    print("エラーが発生しました:", e)
    print("変数が定義されていません。", e)
#エラーがない時に実行はelse
else:
    print("エラーは発生しませんでした。")
#エラーの有無に関わらず実行はfinally
z=3

try:
    result = x / z
except ZeroDivisionError as e:
    result = None
    print("エラーが発生しました:", e)
    print("0で割ることはできません。")
except Exception as e:
    result = None
    print("エラーが発生しました:", e)
    print("変数が定義されていません。", e)
else:
    print("エラーは発生しませんでした。")
finally:
    print("これはエラーの有無に関わらず実行されます。")
    
print("結果:", result)
```

## システム終了以外の全てのエラーをハンドリング

```python
try:
    result = x / y
except ZeroDivisionError as e:
    result = None
    print("エラーが発生しました:", e)
    print("0で割ることはできません。")

else:
    print("エラーは発生しませんでした。")
finally:
    print("これはエラーの有無に関わらず実行されます。")
    
print("結果:", result)
```

##　Webスクレイピングでのエラー処理

```python
import requests
```

```python
url = "https://api.example.com/data"
response = requests.get(url)
print("ステータスコード:", response.status_code)

response.raise_for_status()  
```

```python
url = "https://api.example.com/data"

try:
    response = requests.get(url)
    response.raise_for_status()
except requests.exceptions.HTTPError as e:
    print("HTTPエラーが発生しました:", e)
except requests.exceptions.RequestException as e:
    print("リクエストエラーが発生しました:", e)
else:
    print("リクエストは成功しました。")
finally:
    print("リクエスト処理が終了しました。")
```
